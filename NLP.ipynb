```python
import nltk
nltk.download('wordnet') # wordnet -> database of english
nltk.download('omw-1.4') # open multilingual wordnet, verson 1.4
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
words = ["studies", "running", "better", "went", "files", "mice"]
```

```python
for word in words:
  print(lemmatizer.lemmatize(word))
```

```python
print(lemmatizer.lemmatize("running", pos='v')) # v-> verb, pos -> part of speech
print(lemmatizer.lemmatize("better", pos='a')) # a-> adjective
print(lemmatizer.lemmatize("went", pos='v')) # n-> verb
```

```python
from nltk.corpus import wordnet
```

```python
## tokenization, postag, lemmatized

text = "The cats were running quickly and playing better than before"

# Tokenize
tokens = word_tokenize(text)
print("Tokens", tokens)
```

```python
# pos_tag
tags = pos_tag(tokens)
print("POS Tagging", tags)
```

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.metrics.paice import get_words_from_dictionary
# lemmatize using pos
lemmatized = [lemmatizer.lemmatize(w, pos=get_wordnet_pos(t)) for w, t in tags]
print("Lemmatized", lemmatized)
```

```python
# Remove stop words
final_tokens = [word for word in lemmatized if word.lower() not in stop_words]
print(final_tokens)
```

## Bag-of-words(Bow) -> counts numbers of words and convert it into vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "cats and dogs are friends"
]

vectorizer = CountVectorizer()
bow_matrix = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(bow_matrix.toarray())
```
